```python
#Cleaning
```

```python
#Missing Values
```

```python
df_.loc[df_["TransactionID"].isna(), "TransactionID"] = df_["TransactionID"].mean()
df_.loc[df_["ClientID"].isna(), "ClientID"] = df_["ClientID"].mean()
df_.loc[df_["BirthYear"].isna(), "BirthYear"] = df_["BirthYear"].mean()
df_.loc[df_["Amount"].isna(), "Amount"] = df_["Amount"].mean()
df_.loc[df_["Department"].isna(), "Department"] = df_["Department"].mean()


#df["Age_inputed"] = df["Age"].fillna(df["Age"].mean())
```

```python
df_
```

```python
#Change value-type
```

```python
df_["TransactionID"] = df_["TransactionID"].astype('int')
df_["ClientID"] = df_["ClientID"].astype('int')
df_["BirthYear"] = df_["BirthYear"].astype('int')
df_["Amount"] = df_["Amount"].astype('int')
df_["Department"] = df_["Department"].astype('int')

df_
```

```python
#df_.loc[df_["Profession"]] = df_["Profession"].capitalize()
#df_.loc["Profession"] = df_["Profession"].str.capitalize()
df_
```

```python
# Filling categorical data columns
```

```python
#pd.get_dummies(df_["Profession"])
imputer = KNNImputer(n_neighbors=5)
df_["Profession"] = imputer.fit_transform(df_[["Profession"]])
```

```python
#df_["Profession"] = imputer.fit_transform(df_[["Profession"]])
#df_['Profession'] = df_['Profession'].fillna(df_['Profession'].value_counts().index[0])
df_
```

```python
values = df_["Profession"].unique()
df_["Profession"].map({val:val[0].upper() for val in values})
```

```python
df_['Profession'] = df_['Profession'].str.upper()
df_
```

```python
df_['Profession'] = df_['Profession'].str.lower()
df_
```

```python
df_['Profession'] = df_['Profession'].str.capitalize()
df_
```

```python
df_["Profession"] = df_["Profession"].str.replace('Hr', 'HR', regex=True)
df_
```

```python
df_["Profession"] = df_["Profession"].str.replace('Bdm', 'BDM', regex=True)
df_
```
