# StackAPI

#### Import the necessary libraries here:

```python
# your code here
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
comments = SITE.fetch('comments')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here

questions = SITE.fetch('questions', fromdate=1681406117, todate=1683998117)
questions
```

```python
answers = SITE.fetch('answers', fromdate=1681406117, todate=1683998117)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
question2 = SITE.fetch('questions', fromdate=1683998117,min=5,tagged='python',sort='votes')
question2
```
